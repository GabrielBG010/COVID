```python
###
###  LOAD OF THE INFORMATION IN A DATASET BY STATE
###

import pandas as pd

path = "./flu_ground_truth"

import os

folders = []
state = {}

df = pd.DataFrame(columns=['time']).set_index('time')

# r=root, d=directories, f = files
for r, d, f in os.walk(path):
     for file in f:
            state[file.split('_')[2].split('.')[0]] = pd.read_csv(os.path.join(path,file))
for st_name,df_state in state.items():
    df = pd.concat([df, df_state.rename(columns={'wili': st_name}).set_index('time')],axis=1)

df.reset_index(inplace=True)

for idx in range(len(df['index'])):
    if(len(df['index'][idx])<7):
        df['index'][idx] = df['index'][idx][0:5]+'0'+df['index'][idx][-1]
        
df=df.sort_values(by='index',ascending=True)
df = df.set_index('index')
```
